```python
#necessary libraries

import pandas as pd
import requests
import re
```

Extracting the data from CSV

```python
path_books = '/content/drive/MyDrive/BX-Books.csv'
path_reviews = '/content/drive/MyDrive/BX-Book-Ratings.csv'
path_users = '/content/drive/MyDrive/BX-Users.csv'

reviews = pd.read_csv(path_reviews, encoding= 'unicode_escape', on_bad_lines='skip', sep= ';')
books = pd.read_csv(path_books, encoding= 'unicode_escape', on_bad_lines='skip', sep= ';')
users = pd.read_csv(path_users, encoding= 'unicode_escape', on_bad_lines='skip', sep= ';')
```

Transforming and cleaning the data

```python
#merge reviews and users dataframes
books_dataset = pd.merge(books, reviews, on = 'ISBN')
display(books_dataset)
```

```python
#drop duplicated values
books_dataset = books_dataset.drop_duplicates()
len(books_dataset)
```

```python
#drop null values
books_dataset = books_dataset.dropna()
len(books_dataset)
```

```python
#drop ratings with 0
books_dataset = books_dataset.drop(books_dataset[books_dataset['Book-Rating'] == 0].index)
len(books_dataset)
```

```python
#computing for each book, their average rating and number of ratings
avg_ratings = books_dataset.groupby(['ISBN'])['Book-Rating'].mean().reset_index(name = "Average rating").round(2)
num_ratings = books_dataset.groupby(['ISBN'])['ISBN'].count().reset_index(name = "Number of ratings").round(2)

info_ratings = pd.merge(num_ratings, avg_ratings, left_on = 'ISBN', right_on = 'ISBN').sort_values(by = ['Number of ratings', 'Average rating'], ascending=[False, False])
display(info_ratings)
```

```python
#only taking into consideration the books that have more than 50 ratings
info_ratings = info_ratings[info_ratings['Number of ratings'] >= 50]
display(info_ratings)
```

```python
#adding the average rating and number of ratings in the books dataset by merging the two dataframes
books_dataset = pd.merge(books_dataset.drop(['User-ID','Book-Rating'], axis=1).drop_duplicates(), info_ratings, on = 'ISBN')
display(books_dataset)
```

Extracting data from API


*   for book details I used Google Books API
*   for author details I used Open Library Authors API



```python
#function that makes a list out of the values from a column
#useful for the api call
def list_from_column(dataframe, column):
  column_list = dataframe[column].values.tolist()
  return column_list
```

```python
#function for detting dataframes from API calls
def dataframe_from_api(url, api, check_field, list_items):
  results = [] #list to store results
  
  for item in list_items:
    if(api == 'book'):
      params = {'q': f'isbn:{item} '}
    if(api == 'author'):
      params = f'q={item}'
    
    response = requests.get(url, params=params) # send the GET request to the API and fetch the response
    data = response.json()  # extract the information from the response

    if len(data.get(check_field, [])) > 0: #check if items were returned
      needed_item = data[check_field][0] #only extract the first item
      
      if(api == 'book'): #special case for the book api
        needed_item['ISBN'] = item
      
      results.append(needed_item)
      
  return pd.json_normalize(results)
```

```python
#function for renaming columns
def rename_columns(dataframe):
  new_columns = []
  for column in dataframe.columns:
    new_column = re.sub(r'[_-]', ' ', column).title().replace(' ', '')
    new_columns.append(new_column)
    
  dataframe.columns = new_columns
  return dataframe
```

##### Playing with the datasets

```python
info_dataset = pd.merge(books_dataset, reviews, left_on = 'Isbn', right_on = 'ISBN')
info_dataset = pd.merge(info_dataset, users, on = 'User-ID')
display(info_dataset)
```

```python
#drop ratings with 0
info_dataset = info_dataset.drop(info_dataset[info_dataset['Book-Rating'] == 0].index)
display(info_dataset)
```

```python
info_dataset[['City', 'Region', 'Country']] = info_dataset['Location'].str.split(',', 2, expand=True)
info_dataset['Country'] = info_dataset['Country'].str.strip()

info_dataset_new = info_dataset[['Isbn', 'Country', 'Book-Rating','Age']]
```

```python
print(books_dataset.columns.tolist())
```

```python
def merge_with_books_dataframe(dataset):
  return pd.merge(dataset, books_dataset, on = 'Isbn')[['BookTitle', 'BookDescription', 'NumberOfPages', 'BookAuthor', 'YearOfPublication', 'Publisher', 'ImageUrlM', 'AverageRating', 'NumberOfRatings']]
```

```python
#function that gets the top 10 books that have the highets average rating per country
def best_rated_books_country(dataset, country):
  reviews_country = dataset[dataset['Country'] == country]
  reviews_country = reviews_country.groupby(['Isbn'])['Book-Rating'].mean().reset_index(name = 'Average rating').nlargest(10, 'Average rating')
  return reviews_country

display(merge_with_books_dataframe(best_rated_books_country(info_dataset_new, 'netherlands')).sort_values(by = 'AverageRating', ascending = False))
```

```python
#top 10 authors with books that have the highest ratings

author_sum = info_dataset.groupby(['BookAuthor'])['Book-Rating'].sum().reset_index(name = 'Sum ratings')
author_count = info_dataset.groupby(['BookAuthor'])['BookAuthor'].count().reset_index(name = 'Total ratings')

authors_top = pd.merge(author_sum, author_count, on = 'BookAuthor')

authors_top['Average rating'] = authors_top['Sum ratings'] / authors_top['Total ratings']

display(authors_top.sort_values(by = 'Average rating', ascending = False).head(10))
```
